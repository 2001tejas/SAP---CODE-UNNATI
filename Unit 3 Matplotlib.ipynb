```python
import numpy as np

x = np.arange(0,10,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.arange(0,50,0.10)
plt.xlabel('X axis')
y = np.sin(x)
plt.ylabel('Y axis')
plt.title('Sin graph')

plt.plot(x,y)
```

# subploting

```python
#plot 1
x = (0,1,2,3)
y = (3,8,1,0)
# plot 2
z = (0,1,2,3)
w = (10, 20, 30, 40)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.plot(z,w)

```

```python
#plot 1
x = (0,1,2,3)
y = (3,8,1,0)
#plot 2
z = (0,1,2,3)
w = (10,20,30,40)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(z,w)
plt.subplot(2,2,3)
plt.plot(y,w)
```

```python
# using marker
x = (0,1,2,3)
y = (3,8,1,0)
plt.plot(x,y, marker='*', ms = 20)
```

```python
# using dotted line
x = (0,1,2,3)
y = (3,8,1,0)
plt.plot(x,y, marker='*', ms = 20 , linestyle='dotted', color='g')
         
```

```python
# using bar chart 
x = (3,4,5)
plt.xlabel('values')
y = (2,5,6)
plt.ylabel('sales')
plt.title('my first bar gragh')
plt.bar(x,y)
```

```python
# creating data set
cars=['bmw','tesla','audi', 'jaguar']
data=[23,17,35,44]
plt.pie(data,labels=cars)
plt.plot()
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()

#syntax for 3 d projection
ax = plt.axes(projection = '3d')

# definingg all 3 axis
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

# plotting
ax.plot3d(x,y,z,'r')
ax.set_title('3d plot')
plt.show()
```
